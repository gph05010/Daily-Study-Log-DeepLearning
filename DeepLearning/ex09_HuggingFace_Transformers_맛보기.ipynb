| 방법              | 장점                                         | 단점                                             |
| --------------- | ------------------------------------------ | ---------------------------------------------- |

| 방법              | 장점                               | 단점                                 |
| --------------- | -------------------------------- | ---------------------------------- |
| **노코드 모델링**     | - 접근성<br>- 빠른 프로토타이핑<br>- 직관적 사용 | - 제한된 기능<br>- 유연성 부족<br>- 성능 저하 가능 |

> Make an AI from scratch vs Transfer learning

| 방법                          | 장점                                            | 단점                                  |
| --------------------------- | --------------------------------------------- | ----------------------------------- |

3. 번역하기

```python
# 한글을 영어로 번역하기
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
# AutoTokenizer : 자연어를 컴퓨터가 이해할 수 있도록 벡터로 변환해주는 도구
# AutoModelForSeq2SeqLM : 문장을 입력받아 문장을 출력하는 모델 불러오기
```

```python
model_name = 'Helsinki-NLP/opus-mt-ko-en'
tokenizer = AutoTokenizer.from_pretrained(model_name) # 전용 토크나이저 객체 생성
model = AutoModelForSeq2SeqLM.from_pretrained(model_name) # 학습된 번역 모델 객체 생성
```

```python
# 번역 함수 정의하기
# 한글 텍스트 입력 -> 번역 -> 디코딩
def translate_ko_en (text) :
  input = tokenizer(text, return_tensors='pt') # 입력 텍스트 인코딩(벡터화) pytorch
  output = model.generate(**input) # 번역, ** : 모델에게 이게 텍스트 데이터임을 알려줌
  return tokenizer.decode(output[0], skip_special_tokens = True) # 번역 결과 디코딩, 스페셜 토큰(<pad>, </s> : 모델에게 텍스트의 시작과 끝을 알려줌) 출력하지 않기
```

```python
# 함수 실행
translate_ko_en('안녕하세요 만나서 반갑습니다')
translate_ko_en('혼저옵서예')
```

### 제로샷 분류
- 학습할 때 한 번도 본 적 없는 레이블(클래스)에 대해, 사전 학습된 모델이 알아서 의미적 분류 수행
- 사전 학습된 모델이 새로운 클래스에 대한 학습을 하지 않고, 분류 수행
- 장점 :
  - 학습 데이터가 없거나 부족한 경우 사용
  - 유연하게 레이블을 추가, 변경 가능
- 단점 :
  - 모델이 다루지 못하는 복잡한 데이터의 경우 성능이 저하

```python
# 모델 로딩
# 자연어 학습을 통해 문장 간의 관계를 학습한 모델 -> 우리가 분류하고자 하는 클래스를 직접적으로 학습한 것은 아님
zeroshot_model = pipeline(task = 'zero-shot-classification', model = 'joeddav/xlm-roberta-large-xnli', framework = 'pt')
```

```python
# 분류할 문장
seq = [
    # 드라마 리뷰
    "레이디두아에 나오는 배우 신혜선 연기 진짜 잘하더라 우리엄마가 푹빠져서 살고있어",
    # 정치관련 뉴스 내용
    "최근 선거 후보에 대한 논쟁이 뜨겁습니다. 중앙선거관리위원회는 대부분의 후보가 오늘자로 후보자 등록을 마쳤다고 전했습니다."
]
```

```python
# 후보 레이블 목록
can_labels = ['드라마/영화', '정치', '사회', '경제', '음식', '스포츠']
```

```python
# 모델에게 분류 진행
result = zeroshot_model(seq, # 분류할 문장 리스트
                        can_labels, # 예측할 레이블 리스트
                        multi_label = True) # True 다중레이블, False 단일레이블
```

```python
result[0].keys()
```

```python
for text in result :
  print(f"입력문장 : {text['sequence']}")
  print(f"카테고리 : {text['labels'][0]}")
  print(f"확신도 : {text['scores'][0]}")
  print()
```
